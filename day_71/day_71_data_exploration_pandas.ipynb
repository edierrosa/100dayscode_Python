```python
high_mid_salary = clean_df['Mid-Career Median Salary'].idxmax()
print(f"College major highest mid salary: {clean_df['Undergraduate Major'][high_mid_salary]}")
print(f"Salary: {clean_df['Mid-Career Median Salary'][high_mid_salary]}")
```

```python
low_starting_salary = clean_df['Starting Median Salary'].idxmin()
print(f"College major lowest starting salary: {clean_df['Undergraduate Major'][low_starting_salary]}")
print(f"Salary: {clean_df['Starting Median Salary'][low_starting_salary]}")
```

```python
lowest_mid_salary = clean_df['Mid-Career Median Salary'].idxmin()
print(f"College major lowest mid salary: {clean_df['Undergraduate Major'][lowest_mid_salary]}")
print(f"Salary: {clean_df['Mid-Career Median Salary'][lowest_mid_salary]}")
```

```python
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
new_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', new_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
largest_difference = clean_df.sort_values('Spread', ascending=False)
largest_difference[['Undergraduate Major', 'Spread']].head()
```

```python
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
mid_career_highests = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
mid_career_highests[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```
